**Q3. What are the different ways to define and call a function in Python?**

```python
# 1. Regular Function
# Definition:
def greet(name):
    return f"Hello, {name}!"
# Calling:
print(greet("Alice"))
```

```python
# 2. Lambda Function
# Definition:
greet = lambda name: f"Hello, {name}!"
# Calling:
print(greet("Alice"))
```

```python
# 4. Function with Variable-Length Arguments
# Definition:
def greet(*names):
    return [f"Hello, {name}!" for name in names]
# Calling:
print(greet("Alice", "Bob", "Charlie"))
```

```python
# 5. Function with Keyword Arguments
# Definiton:
def greet(**kwargs):
    return f"Hello, {kwargs.get('name', 'Guest')}!"
# Calling:
print(greet(name="Alice"))
```

**Q5.  What are iterators in Python and how do they differ from iterables?**

--> In Python, iterators and iterables are closely related concepts, but they have distinct roles. Let's break down the differences:

Iterables

Definition: An iterable is any Python object capable of returning its members one at a time, allowing it to be iterated over in a loop. Examples include lists, tuples, strings, and dictionaries.

How to Identify: An object is iterable if it implements the __iter__() method, which returns an iterator.

Iterators

Definition: An iterator is an object that represents a stream of data. It returns the next item in the sequence when you call the __next__() method. Iterators are used to iterate over iterables.

How to Identify: An object is an iterator if it implements both the __iter__() and __next__() methods.

Key Differences

Creation:

Iterable: Created using collection data types like lists, tuples, strings, etc.

Iterator: Created using the iter() function on an iterable.

Methods:

Iterable: Must implement the __iter__() method.

Iterator: Must implement both __iter__() and __next__() methods.

Usage:

Iterable: Can be used in a for loop directly.

Iterator: Used to fetch items one at a time using the next() function.

```python
my_list = [1, 2, 3, 4, 5]
my_iterator = iter(my_list)
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
```

**Q6.  Explain the concept of generators in Python and how they are defined.**

--> Generators in Python are a special type of iterable that allow you to iterate over a sequence of values lazily, meaning they generate values on the fly and only when needed. This makes them memory-efficient, especially when dealing with large datasets.

Defining Generators

Generators are defined using functions and the yield statement. Unlike regular functions that use return to return a value and terminate, a generator function uses yield to return a value and pause its execution, maintaining its state for subsequent calls.

```python
def fibonacci_sequence(n):
    a, b = 0, 1
    count = 0
    while count < n:
        yield a
        a, b = b, a + b
        count += 1

for number in fibonacci_sequence(10):
    print(number)
```

**Q7. What are the advantages of using generators over regular functions?**

--> Generators offer several advantages over regular functions, especially when dealing with large datasets or streams of data. Here are some key benefits:

1. Memory Efficiency

Generators do not store the entire sequence in memory; they generate each value on the fly. This makes them ideal for working with large datasets or infinite sequences.

2. Lazy Evaluation

Generators produce values only when requested, which can improve performance and reduce the initial load time.

3. Simplified Code

Generators can simplify code by replacing complex iterator classes with a simple function using yield.

4. Improved Performance

Generators can improve performance by avoiding the overhead of creating and storing large data structures.

```python
def generate_numbers(n):
    for i in range(n):
        yield i

# Using the generator
for number in generate_numbers(10):
    print(number)
```

```python
def fibonacci_sequence():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Using the generator
fib_gen = fibonacci_sequence()
for _ in range(10):
    print(next(fib_gen))
```

```python
def squares(n):
    for i in range(n):
        yield i ** 2

# Using the generator
for square in squares(10):
    print(square)
```

**Q8. What is a lambda function in Python and when is it typically used?**

--> A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike regular functions defined with def, lambda functions can have any number of arguments but only one expression. The expression is evaluated and returned when the function is called.

```python
# Example
add = lambda x, y: x + y
print(add(2, 3))
```

```python
# Typical Uses :-
# Short, Simple Functions
square = lambda x: x ** 2
print(square(4))
```

```python
# Higher-Order Functions
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)
```

```python
# Sorting and Key Functions
my_list = [(1, 3), (2, 1), (3, 2)]
sorted_list = sorted(my_list, key=lambda x: x[1])
print(sorted_list)
```

```python
# Inline Functions
def apply_function(func, value):
    return func(value)

result = apply_function(lambda x: x ** 3, 3)
print(result)
```

**Q9. Explain the purpose and usage of the `map()` function in Python.**

-->The map() function in Python is used to apply a given function to all items in an iterable (such as a list or tuple) and return a map object (an iterator) containing the results. It is particularly useful for transforming data in a concise and readable way.

Purpose

The primary purpose of map() is to simplify the process of applying a function to each item in an iterable, avoiding the need for explicit loops.

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
celsius_temps = [0, 20, 37, 100]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))
print(fahrenheit_temps)
```

**Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

--> Key Differences

Functionality:

map(): Transforms each item in the iterable.

reduce(): Aggregates the iterable into a single value.

filter(): Selects items from the iterable based on a condition.

Return Type:

map(): Returns a map object (an iterator).

reduce(): Returns a single value.

filter(): Returns a filter object (an iterator).

Use Cases:

map(): When you need to apply a function to each item in an iterable.

reduce(): When you need to combine all items in an iterable into a single value.

filter(): When you need to select items from an iterable based on a condition.

```python
# map ()
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)
```

```python
# reduce ()
from functools import reduce

numbers = [1, 2, 3, 4, 5]
sum_of_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_of_numbers)
```

```python
# filter ()
numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):

    even_sum = 0

    for number in numbers:
        if number % 2 == 0:
            even_sum += number

    return even_sum

numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(numbers)
print(result)
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True


for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number")
    else:
        print(f"{num} is not a prime number")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.current_term = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term >= self.num_terms:
            raise StopIteration
        if self.current_term == 0:
            self.current_term += 1
            return self.a
        elif self.current_term == 1:
            self.current_term += 1
            return self.b
        else:
            self.current_term += 1
            self.a, self.b = self.b, self.a + self.b
            return self.b

num_terms = 10
fib_iterator = FibonacciIterator(num_terms)

for fib_number in fib_iterator:
    print(fib_number)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

exponent = 5
power_generator = power_of_two(exponent)

for power in power_generator:
    print(power)
```

```python
# 9. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_string))

my_string = "Hello, World!"
print(remove_vowels(my_string))
```
